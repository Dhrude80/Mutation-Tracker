```python
import random
import matplotlib.pyplot as plt
import numpy as np

genes = ["A", "T", "G", "C"]
DNA = []


for x in range(10):
    original_dna = random.choice(genes)
    DNA.append(original_dna)

each_mutation = np.empty((0,len(DNA)))
plot_mutation = np.empty((1000, len(DNA)))
number_of_mutation_changes =  []

for times in range(1000):
    for i in range(len(DNA)):
        dna_str = DNA[i]
        chance = random.random()
        if chance <= 0.02:
            new_dna_str = random.choice(genes)
            if new_dna_str != dna_str:
                DNA[i] = new_dna_str
    each_mutation = np.append(each_mutation, [list(DNA)], axis = 0)
            
for row_index , row in enumerate(each_mutation):
    for col_index, val in enumerate(row):
        if val == "A":
            each_mutation[row_index][col_index] = 1
        elif val == "T":
            each_mutation[row_index][col_index] = 2
        elif val == "G":
            each_mutation[row_index][col_index] = 3
        else:
            each_mutation[row_index][col_index] = 4
    plot_mutation = np.append(plot_mutation, [list(row)], axis = 0)

for row_i in range(1, len(plot_mutation)):
    difference = 0
    for i in range(len(plot_mutation[row_i])):
        if plot_mutation[row_i][i] != plot_mutation[row_i-1][i]:
            difference += 1
    number_of_mutation_changes.append(difference)

plt.plot(number_of_mutation_changes)
plt.xlabel("Generation")
plt.ylabel("Mutation change vs previous generation")
plt.show()
```
